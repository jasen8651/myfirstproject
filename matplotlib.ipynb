```python
#그래프

fig=plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1) #전체 장을 가로 1한 세로 2칸으로 나눈 다음 1번칸에 배치  ax1의  axes생성
ax1.set(xlim=[0,30], ylim=[0,30], title= 'AX1 Example', xlabel= 'xAxis', ylabel='yAxis') #범위 지정 및 이름 지정
x=np.array([1,12,25,70])
y=np.array([4,10,5,12])
ax1.plot(x,y, color='red')


ax2= fig.add_subplot(1,2,2) #전체 창을 가로 1한 세로 2칸 나눈후 2번칸에 배치 ax2의  axes생성
ax2.set_xlim([0,30])
ax2.set_ylim([0,30])
ax2.set_title('ax2 example', size=30)
ax2.set_xlabel('xaxis', size =15)
ax2.set_ylabel('yaxis', size =15)
ax2.scatter(x,y, marker='+')
plt.show()
```
