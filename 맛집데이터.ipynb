```python
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
import os

from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.keys import Keys

import pandas as pd # csv -> dataframe으로 전환
import numpy as np  # numpy 패키지 가져오기

#selenium V3
#path = './lib/chromedriver.exe
#driver.get(path)

# selenium V4
#chrome_options = webdriver.ChromeOptions()
#driver = webdriver.Chrome(service=Service(ChromeDriverManger().install()))

#Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
print(dir(driver))
```

```python
seoulfood = pd.read_csv('pdsample/seoulfood.csv', encoding='cp949')
# print(seoulfood)
# seoulfood_size = seoulfood['foodstore_id'].size

seoulfood_states1 = seoulfood[seoulfood['area'] == "자곡동"]
print(dir(np))
type(seoulfood)
print(dir(seoulfood))
seoulfood_copy = seoulfood_states1.copy()
# print(seoulfood_copy)
dict_seoulfood = seoulfood_states1.to_dict()

print(dict_seoulfood['foodstore_id'])
```

```python
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException

# shop_name = [dict_seoulfood['foodstore_id'][index] for index in range(seoulfood_size)]
shop_name = [seoulfood_states1['foodstore_id'][index] for index in range(seoulfood_states1_size)]
shopcode = {}
for s_name in shop_name:

  url = 'https://map.kakao.com/'
  driver.get(url)

  # 크롤링 데이터 확인
  print(type(driver))
  #print(driver.page_source)
  print(type(driver.page_source))

  # find_element -> WebElement으로 리턴
  inputSearch = driver.find_element(By.CSS_SELECTOR, '#search\.keyword\.query')
  #search\.keyword\.query
  print(inputSearch)

  # 검색어를 입력한다.
  inputSearch.send_keys(s_name)
  time.sleep(1)

  # id = 'search_btn' 인 검색버튼을 찾는다.
  buttonSearch = driver.find_element(By.CSS_SELECTOR, '#search\.keyword\.submit')
  buttonSearch.send_keys(Keys.ENTER)
  print(buttonSearch.text)

  time.sleep(1)


  time.sleep(1)

  try :
    menu_list_a = driver.find_element(By.XPATH, '//*[@id="info.search.place.list"]/li[1]/div[5]/div[4]/a[1]').get_attribute("href")

    # list 형식
    # shopcode.append(menu_list_a.split("/")[3])

    # dictionary 형식
    shopcode[s_name] = menu_list_a.split("/")[3]

    print(shopcode)
  except NoSuchElementException :
    # list 형식
    # shopcode.append(None)

    # dictionary 형식
    shopcode[s_name] = None

    print(shopcode)

    #info\.search\.place\.list > li:nth-child(1)
    #info\.search\.place\.list > li:nth-child(1) > div.head_item.clickArea > strong > a.link_name

    # csv 파일로 저장


#seoulfood_states1.to_csv("pdsample/seoulfood_양천구", encoding='cp949')
# f = open(f'{"seoulfood_양천구"}.csv','w',encoding='utf-8',newline='') #파일오픈
#  csvWriter = seoulfood.csv.writer(f)#열어둔 파일
#  for i in seoulfood_states1:
#    csvWriter.writerow(i)
#  f.close()
#  print("완료"); 
```
